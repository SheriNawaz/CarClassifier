```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("prondeau/the-car-connection-picture-dataset")

print("Path to dataset files:", path)
```

```python
import glob

image_paths = glob.glob(f"{path}/**/*.jpg", recursive=True) + \
              glob.glob(f"{path}/**/*.png", recursive=True)

print(f"Found {len(image_paths)} images")
for img_path in image_paths[:5]:  # print a few sample paths
    print(img_path)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(image_paths[0])
plt.imshow(img)
plt.axis('off')
plt.title("Sample Image")
plt.show()
```
