# Importing Dataset from https://www.kaggle.com/datasets/prondeau/the-car-connection-picture-dataset



```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("prondeau/the-car-connection-picture-dataset")

print("Path to dataset files:", path)
```

# Create list of file paths

```python
import glob

image_paths = glob.glob(f"{path}/**/*.jpg", recursive=True) + \
              glob.glob(f"{path}/**/*.png", recursive=True)

print(f"Found {len(image_paths)} images")
```

# Convert the list into a Panda dataframe and label each image

```python
import os
import pandas as pd

data = []

for path in image_paths:
  filename = os.path.basename(path)
  parts = filename.split('_')
  data.append({"Path": path, "Name": f"{parts[0]} {parts[1]} {parts[2]}"})

df = pd.DataFrame(data)
print(df.head())
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(df.iloc[1345, 0])
plt.imshow(img)
plt.axis('off')
plt.title(df.iloc[1345, 1])
plt.show()
```
